```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
    
%matplotlib inline
```

```python
def read_auto_data(fileName = "Automobile price data.csv"):
    'Function to load the auto price data set from a .csv file' 
    import pandas as pd
    import numpy as np

    ## Read the .csv file with the pandas read_csv method
    data = pd.read_csv(fileName)
    
    ## Remove rows with missing values, accounting for mising values coded as '?'
    cols = ['price', 'bore', 'stroke', 
          'horsepower', 'peak-rpm']
    for column in cols:
        data.loc[data[column] == '?', column] = np.nan
    data.dropna(axis = 0, inplace = True)

    ## Convert some columns to numeric values
    for column in cols:
        data[column] = pd.to_numeric(data[column])
        
    return data
data = read_auto_data()
```

### Типы графиков

- **Line plot** - График функции
- **Bar plot** - Столбчатая диаграмма 
- **Histogram** - Гистограмма 
- **Heat Map** - Корреляционная матрица
- **Box plot** - Боксплот (т.н. "ящик с усами") -
- **KDE** - Kernel Density Estimator - Ядерная сглаживание для оценки плотности 
- **Violin plot** - Диаграмма в виде скрипки
- И их комбинации

### Scatter plot

```python
fig, ax = plt.subplots(figsize=(10, 10))

plt.scatter( data['price'], data['horsepower'], marker='o', 
            alpha=0.5, s=data['engine-size']*1.3, cmap = 'paired', label='Машина')
ax.set(xlabel='Цена', ylabel='Лошадиные силы', title='Диаграмма рассеяния цена-лошадиные силы',
       xlim=(data['price'].min()*0.85, data['price'].max()*1.02), 
       ylim = (data['horsepower'].min()*0.95, data['horsepower'].max()*1.02))
plt.grid()
plt.legend(loc='upper left', prop={'size':12})

plt.show()
```

### Line plot

```python
N = 100
a = np.random.randn(N) 
b = (np.random.randn(N+50))
x1 = range(0, 100)
x2 = range(0, 150)

fig, ax = plt.subplots(2, 1, figsize = (8, 5))
fig.subplots_adjust(hspace=0.5)

ax[0].plot(x1, a)
ax[0].set(xlim=(0, 99), xlabel=r'$Days \to Weeks$', ylabel='Value')
ax[0].grid()

ax[1].plot(x2, b, color='orange')
ax[1].set(xlim=(0, 150), xlabel=r'$time, \mu s$')
ax[1].set_ylabel(r'$\frac{\omega*\delta}{\beta^4}$', rotation=45, fontsize=20)
ax[1].grid(True)

plt.show()
```

### Bar plot

```python
fig, ax = plt.subplots(figsize=(8, 4))

sns.countplot(x='body-style', hue = 'aspiration',data=data, palette='Set2', 
              order = data['body-style'].value_counts().index)
plt.xticks(rotation=30)

plt.show()
```

### Histogram

```python
fig, ax = plt.subplots(figsize=(8, 5))

sns.distplot(data['length'], ax=ax, kde=True)
ax.set(xlabel='Длина машины', ylabel='Кол-во машин', 
       xlim=(data['length'].min()*.99, data['length'].max()*1.01))

plt.show()
```

### Heat map

```python
sample_data = data[['length', 'engine-size', 'horsepower', 'city-mpg', 'price']].copy()
corr = sample_data.corr()
```

```python
sns.heatmap(corr)
```

```python
def nice_heatmap(data, size=None):
    corr = data.corr(method='pearson')
    mask = np.zeros_like(corr, dtype=np.bool)
    
    f, ax = plt.subplots(figsize=size)
    sns.heatmap(corr, mask=mask, vmax=1, center=0, annot=True,
                fmt='.3f', square=True, linewidths=.5, cbar_kws={'shrink':.5})
    plt.yticks(rotation=0)
    
    plt.show()
    
nice_heatmap(sample_data, (8, 8))
```

### Violin plot

__Violin plot = Box plot + KDE__

```python
fig, ax = plt.subplots(figsize=(8, 8))
sns.violinplot(x = 'body-style', y = 'horsepower',
              data=data, palette='Set2')
ax.set(xlabel=('Тип кузова'), ylabel=('Лошадиные силы'))
plt.show()
```
